Our dataset contains weekly retail prices of three grades of gasoline, diesel, auto propane and compressed natural gas in 10 Ontario markets. It is downloaded from https://data.ontario.ca/dataset/fuels-price-survey-information.


```python
import pandas as pd
import numpy as np
df = pd.read_csv('data/fueltypesall.csv')
```

Mean Regular Unleaded Gasoline price among different cities in 2019

```python
year_condition = pd.to_datetime(df['Date']).dt.year == 2019
fuel_condition = df["Fuel Type"] == "Regular Unleaded Gasoline"
temp = df.loc[fuel_condition & year_condition].mean()
graph = temp[0:10].plot.bar(title = "Regular Unleaded Gasoline price among different cities in 2019")
graph = graph.set(xlabel="Cities", ylabel="Gasoline Prices")
```

Density distribution (with Gaussian kernel) of the Regular Unleaded Gasoline price in Ottawa for 2019

```python
year_condition = pd.to_datetime(df['Date']).dt.year == 2019
fuel_condition = df["Fuel Type"] == "Regular Unleaded Gasoline"
city_condition = 'Ottawa'
temp = df.loc[year_condition & fuel_condition, city_condition]
graph = temp.plot.kde(title = "Density Distribution Regular Unleaded Gasoline in Ottawa for 2019")
graph = graph.set(xlabel='Gasoline Prices')
```

Boxplot for the Regular Unleaded Gasoline price in Ottawa across year 2012-2019

```python
df['Year'] = pd.to_datetime(df['Date']).dt.year
year_lower = df['Year'] >= 2012
year_higher = df['Year'] <= 2019
fuel_condition = df["Fuel Type"] == "Regular Unleaded Gasoline"
temp = df.loc[(year_lower & year_higher) & fuel_condition]
graph = temp.boxplot(column='Ottawa',by='Year')
graph = graph.set(xlabel = 'Years', ylabel='Gasoline Prices')
```
